```python
import pandas as pd
```

```python
labelsM = pd.read_csv("phqsMoodableCommaQuoteWindows.csv")
labelsE = pd.read_csv("phqsEMUCommaQuoteWindows.csv")

idM = []
for l in labelsM["id"]:
    idM.append("m" + str(l))
labelsM["id"] = idM

idE = []
for l in labelsE["id"]:
    idE.append("e" + str(l))
labelsE["id"] = idE

labels = labelsM.append(labelsE)

print(labelsM.shape)
print(labelsE.shape)
print(labels.shape)

labels.head()
```

```python
qs = []
for i in labels.content:
    temp = i[1:-1].split(",")
    scores = []
    for j in temp:
        temp2 = j[-2:-1]
        scores.append(temp2)
    qs.append(scores)

qdf = pd.DataFrame(qs)
qdf["id"] = list(labels["id"])
qdf = qdf.drop_duplicates(subset = "id", keep = "last").reset_index(drop=True)
qdf.head()
```

```python
filenameList = ["callIn14_4", "callOut14_4", "textIn14_4", "textOut14_4", "callIn14_6", "callOut14_6", "textIn14_6", "textOut14_6", "callIn14_12", "callOut14_12", "textIn14_12", "textOut14_12", "callIn14_24", "callOut14_24", "textIn14_24", "textOut14_24", "callAll14_4", "callAll14_6","callAll14_12", "callAll14_24", "textAll14_4", "textAll14_6", "textAll14_12", "textAll14_24"]
filename2List = ["_average_length", "_counts", "_unique_contacts", "_combined"]

for filename in filenameList:
    for filename2 in filename2List:
        df = pd.read_csv("TimeSeries/" + filename + ".csv")
        data = pd.read_csv("TimeSeriesFeatures/" + filename + filename2 + ".csv")
        data['id'] = list(df["id"])
        data = data.drop(columns = ["Unnamed: 0", "depressed"])
        new = data.merge(qdf, on = ['id'])
       
        if data.shape[0] != new.shape[0]:
            print(filename)
        
        new.to_csv("data/" + filename + filename2 + ".csv", index = False)
```
